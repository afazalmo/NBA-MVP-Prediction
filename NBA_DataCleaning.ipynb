## Cleaning the Players Table

```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
del players["Rk"] ## delete rank column as it isn't needed
```

```python
players.head()
```

```python
## remove astericks from player names 
players["Player"] = players["Player"].str.replace("*","",regex=False)
```

```python
players["Player"].head(50)
```

Some of the players have multiple names associated with each year we need to reduce these to one row per year:

```python
## Get the total row only for given player

def single_row(df):
    ## if data frame has only 1 row return
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]  
        row["Tm"] = df.iloc[-1,:]["Tm"]  ## replace "TOT" team name to current team name
        return row
    
## group the player and year and apply the single row function to the dataframe    
players = players.groupby(["Player","year"]).apply(single_row)
```

```python
## drop two extra indeces  
players.index = players.index.droplevel()
players.index = players.index.droplevel()
```

```python
mvps.info()
```

```python
players.info()
```

```python
del players["Unnamed: 0"]
```

## Combining the players and mvp data in to one table

```python
## outer merge to keep players who aren't MVPs
combined = players.merge(mvps,how="outer", on=["Player","year"])
combined.head(5)
```

```python
## replace NaN values with 0
combined[["Pts Won","Pts Max", "Share"]] = combined[["Pts Won","Pts Max", "Share"]].fillna(0)
```

```python
combined
```

## Cleaning the Teams Data

```python
teams = pd.read_csv("teams.csv")
```

```python
teams
```

```python
teams.head(30)
```

```python
del teams["Unnamed: 0"]
```

The teams dataset has rows that repeat division names and is detrimental to the readibility and accesibility of the data. 

We'll get rid of those rows now:

```python
## keep only rows that don't contain "Division"
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams.head(30)
```

```python
## remove astericks from team names
teams["Team"] = teams["Team"].str.replace("*","",regex=False)
```

```python
teams.head(5)
```

```python
teams["Team"].unique()
```

```python
combined["Tm"].unique()
```


When looking at the team names on the combined table and teams table we can see that team names are different(abbreviations 
vs whole names). We'll add a team column in the combined table by mapping values accordingly to solve this issue

```python
## create dictionary to map values 
nicknames = {}

## read in the nicknames csv and remove excess characters 
with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        line = line.replace('"',"")
        abbrev, name = line.replace("\n","").split(",")
        
        nicknames[abbrev] = name ## set abbreviation and name as key/value pairs
```

```python
combined["Team"] = combined["Tm"].map(nicknames) ## create new column Team with mapped values 
```

```python
combined.head(5)
```

## Merging MVP, Player, and Team Tables

```python
## merge combined table with team table on the same team and year
stats = combined.merge(teams,how="outer", on=["Team","year"])
```

```python
stats
```

```python
stats.info()
```


Most of these columns are numeric columns but are stored as strings. Let's convert some of these columns to numeric columns


```python
## Go through every column checks if its convertible if so, convert
stats = stats.apply(pd.to_numeric, errors="ignore")
```

```python
stats.dtypes
```

The "GB" column should be a numeric columns as well, but didn't convert. Let's see why:

```python
stats["GB"].unique()
```

The column contains '-' which is preventing it from being converted in to a numeric column

```python
stats["GB"]=stats["GB"].replace("—","0.0") ## replace all instances of "-" with 0.0
```

```python
stats["GB"] = stats["GB"].apply(pd.to_numeric) ## convert to numeric values again
```

```python
stats.info()
```

```python
stats.to_csv("player_mvp_stats.csv") ## save combined data to csv file
```

## Analyzing the Stats Table

Now that we've cleaned and combined the data, we can use the combined table to extract and dsiplay various stat points throughout the scraped years

Let's start by looking at te highest scoring players

```python
highest_scoring = stats[stats["G"] > 70].sort_values(by="PTS",ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt

highest_scoring.plot.bar("Player","PTS")
plt.xlabel("Player")
plt.ylabel("Points")
plt.title("NBA Top Scorers(1999-2021)")
```

**Let's see the highest scorer for each year:**

```python
## group by year and for every year get the highest scorer
highest_scoring = stats.groupby("year").apply(lambda x: x.sort_values("PTS",ascending=False).head(1))
```

```python
highest_scoring.plot.bar("year","PTS")
plt.xlabel("Year")
plt.ylabel("Points")
plt.title("Highest Point totals Per Year(1999-2021)")
```

Take a look at the correlation data for each column with the MVP vote share:

```python
stats.corr()["Share"]
```

```python
stats.corr()["Share"].plot.bar()
```

Analyzing the correlation data we can see that various columns correlate with the share of the MVP vote. The columns "PTS", "FT", "A" and various other scoring columns have the highest correlation (We ignore "PTS Won", "PTS Max", and "Share" as they are directly related to the votes) and we can see that columns such as "MP" and "W/L%" also determine the share of the vote.
